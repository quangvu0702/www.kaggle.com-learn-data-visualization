```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
flight_delays_data = pd.read_csv('data/flight_delays.csv', index_col='Month')
```

```python
flight_delays_data.head()
```

```python
plt.figure(figsize=(16,8))
for c in flight_delays_data.columns:
    sns.lineplot(data=flight_delays_data[c], label=c)
```

### Bar

```python
plt.title('Average Arrival Delay for Spirit Airlines Flights, by Month')
sns.barplot(x=flight_delays_data.index, y=flight_delays_data.AS)
plt.ylabel('Arrival delay (in minutes)')
```

### Heatmap

```python
plt.figure(figsize=(16,6))
plt.title('Average Arrival Delay for Each Airline, by Month')
sns.heatmap(data=flight_delays_data, annot=True)
plt.xlabel('Airline')
```
